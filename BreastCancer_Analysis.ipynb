```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
# Use labelencoder to convert categorical features to numerical....
le = LabelEncoder()
```

```python
# Use labelencoder to convert categorical features to numerical....

data['Race'] = le.fit_transform(data['Race'])
data['Marital Status'] = le.fit_transform(data['Marital Status'])
data['T Stage '] = le.fit_transform(data['T Stage '])
data['N Stage'] = le.fit_transform(data['N Stage'])
data['6th Stage'] = le.fit_transform(data['6th Stage'])
data['differentiate'] = le.fit_transform(data['differentiate'])
data['A Stage'] = le.fit_transform(data['A Stage'])
data['Tumor Size'] = le.fit_transform(data['Tumor Size'])
data['Estrogen Status'] = le.fit_transform(data['Estrogen Status'])
data['Progesterone Status'] = le.fit_transform(data['Progesterone Status'])
data['Status'] = le.fit_transform(data['Status'])
```

```python
data
```

```python
# New datasets with numerical features.... 
data
```

```python
# Drop less important Features for training and testing.
X=data.drop(columns=['Status','Grade'])
y=data['Status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
#Use Xgboost,accuracy_score to fit and predict model.
xgb = XGBClassifier()
xgb.fit(X_train,y_train)
ypred=xgb.predict(X_test)
accuracy_score(ypred,y_test)
```

```python
#Use Randomforest model,accuracy_score to fit and predict model.

rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
ypred = rfc.predict(X_test)
accuracy_score(ypred,y_test)
```
